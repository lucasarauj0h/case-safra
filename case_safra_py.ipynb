### Case 2

Escreva uma query que encontre os pre√ßos de todos os produtos no dia 2019-08-16

-  Deve-se assumir que os pre√ßos iniciais de todos os produtos antes de qualquer mudan√ßa de pre√ßos √© 10
-  As colunas product_id e change_date formam a chave prim√°ria da tabel (colunas com comina√ß√£o de valores √∫nica)
-  Cada linha da tabela representa a mudan√ßa de pre√ßo de um produto em uma determinada data

```
products = {
    'product_id': [1, 2, 1, 1, 2, 3, 3, 2],
    'new_price': [20, 50, 30, 35, 65, 20, 20, 75],
    'change_date': ['2019-08-14', '2019-08-14', '2019-08-15', '2019-08-16', '2019-08-17', '2019-08-18', '2019-08-21', '2019-08-23']
}
```

```
# Verificando os tipos de dado das vari√°veis/colunas
df_products = pd.DataFrame(products)
df_products.dtypes
```

```
# Alterando tipo da coluna para data
df_products['change_date'] = pd.to_datetime(df_products['change_date'])
```

```
# Query para obter os pre√ßos mais recentes de cada produto antes da data especificada e transofrma-l√° em um DF
ultimos_precos = df_products[df_products['change_date'] <= '2019-08-16'].sort_values(by=['product_id', 'change_date'], ascending=[True, False]).drop_duplicates(subset='product_id').set_index('product_id')['new_price']
```

```
ultimos_precos.head()
```

```
# Query para obter todos os produtos presentes na base de dados
todos_produtos = pd.DataFrame({'product_id': df_products['product_id'].unique()})

# Jun√ß√£o (join) dos dois DFs, contendo os produtos com o valor mais recentes, e aqueles que ainda n√£o haviam atualizado, s√£o inseridos com o valor padr√£o -10-
prices = todos_produtos.join(ultimos_precos, on='product_id')
prices.rename(columns={'new_price': 'price'}, inplace=True)
prices['price'].fillna(10, inplace=True)
prices.head()
```
